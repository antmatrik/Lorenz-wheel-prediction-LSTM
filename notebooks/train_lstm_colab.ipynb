## 5. Save weights into the project

Copy the trained artifacts to `weights/` under the canonical names, then **download
them and commit them to the repo** (`git add weights/ && git commit && git push`) so
the evaluation notebook — and anyone who clones — can use them.

```python
import os, shutil

os.makedirs('weights', exist_ok=True)
pairs = [
    ('outputs/lstm_weights.weights.h5', 'weights/lstm.weights.h5'),
    ('outputs/lstm_stats.npz',          'weights/lstm_stats.npz'),
]
for src, dst in pairs:
    if os.path.exists(src):
        shutil.copy(src, dst); print('saved', dst)
    else:
        print('skip (not found):', src)
```
