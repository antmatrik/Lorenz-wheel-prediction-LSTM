## 5. Transfer weights back (Google Drive)

Colab storage is wiped when the session ends, so copy the artifacts to Drive.
Mount Drive, then copy the `outputs/` files. Afterwards download them from Drive
to your machine's local `outputs/` folder.

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
import os
import shutil

DRIVE_DIR = "/content/drive/MyDrive/lorenz_waterwheel_lstm"
os.makedirs(DRIVE_DIR, exist_ok=True)

for name in ["lstm_weights.weights.h5", "lstm_stats.npz", "lstm_predictions.csv"]:
    src = os.path.join("outputs", name)
    if os.path.exists(src):
        shutil.copy(src, os.path.join(DRIVE_DIR, name))
        print("saved", name, "->", DRIVE_DIR)
    else:
        print("skip (not found):", src)
```
